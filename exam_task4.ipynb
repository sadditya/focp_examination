```python
def all_between_0_and_100(numbers):
    for n in numbers:
        if n < 0 or n > 100:
            return False
    return True

# Test the function
test_list_1 = [10, 25, 50, 99, 100]
test_list_2 = [5, 20, 150, 40]

print("Test list 1:", all_between_0_and_100(test_list_1))   # True
print("Test list 2:", all_between_0_and_100(test_list_2))   # False
```
